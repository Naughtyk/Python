```python
# Импортируем библиотеку
import tkinter as tk 

# Задаём размеры поля
BRD_ROWS = BRD_COLS = 8
# и размер ячейки
CELL_SZ = 50

# Создаём объект класса Tk
root = tk.Tk()

# Создаём фон/поле/холст, на котором будут рисоваться объекты
canvas = tk.Canvas(root, width=CELL_SZ*BRD_ROWS, height=CELL_SZ*BRD_COLS)

# Цвета ячеек - белый и черный
cell_colors = ['white', 'black']
ci = 0  # индекс цвета

flag = False # Флаг для отслеживания, была ли уже запущена анимация или нет

# Рисуем собственно поле
for row in range(BRD_ROWS):
    for col in range(BRD_COLS):
        x1, y1 = col * CELL_SZ, row * CELL_SZ # Координаты верхнего левого угла квадрата
        x2, y2 = col * CELL_SZ + CELL_SZ, row * CELL_SZ + CELL_SZ # Координаты нижнего правого угла квадрата
        canvas.create_rectangle((x1, y1), (x2, y2), fill = cell_colors[ci]) # Сам квадрат

        ci = not ci # Меняем цвет

    ci = not ci

canvas.pack() # Упаковка, чтобы всё было чётко и ровно
shashka = canvas.create_oval(10, 10, 40, 40, outline="gray", fill="brown", width=2) # Создаём шашку

# Функция передвижения шашки по диагонали вниз
def motion():
    canvas.move(shashka, 50, 50)
    if canvas.coords(shashka)[2] < 350:
        root.after(700, motion)
    else:
        root.after(700, backmotion)

# Функция передвижения шашки по диагонали вверх (обратно)
def backmotion():
    canvas.move(shashka, -50, -50)
    if canvas.coords(shashka)[2] > 50:
        root.after(700, backmotion)
    else:
        root.after(700, motion)

# Функция, вызываемая в случае нажатия кнопки "Запуск движения"
def clicked():
    #button.configure(state = 'disabled') # Если хотим реализовать через кнопку
    global flag # Чтобы при повторном нажатии уже ничего не происходило
    if flag == False:
        flag = True
        motion()
    else:
        pass

#button = tk.Button(root, text = "Запуск движения", width = 20, height = 5, command = clicked)
#button.place_configure(x = 400, y = 150)

# Реализация вызова контекстного меню с помощью ПКМ
class Main(tk.Frame):
    def __init__(self, parent):
        tk.Frame.__init__(self, parent)
        self.parent = parent
        self.parent.title("context")
        self.menu = tk.Menu(self.parent, tearoff = 0)
        self.menu.add_command(label = "Запуск движения", command = clicked)
        self.parent.bind("<Button-3>", lambda event: self.menu.post(event.x_root, event.y_root))
        self.pack()

# Создаём объект класса Main, чтобы нажатие по ПКМ вызывало контекстное меню
app = Main(root)

root.mainloop()
```
